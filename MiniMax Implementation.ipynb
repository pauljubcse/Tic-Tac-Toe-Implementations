```python
class TicTacToe:
    def __init__(self, board):
        self.board=board
        self.player='O'
        self.computer='X'


    def run(self):
        print("Computer Starts")

        while True:
            self.moveComputer()
            self.movePlayer()
    
    def printBoard(self):
        print(self.board[1]+'|'+self.board[2]+'|'+self.board[3])
        print('-+-+-')
        print(self.board[4]+'|'+self.board[5]+'|'+self.board[6])
        print('-+-+-')
        print(self.board[7]+'|'+self.board[8]+'|'+self.board[9])
        print('\n')
    
    def isDraw(self):
        #if empty cells are there, cant be draw
        for position in self.board.keys():
            if self.board[position]==' ':
                return False
        return True
    
    def isWinning(self, player):
        #DIAGONALS
        if self.board[1]==player and self.board[5]==player and self.board[9]==player:
            return True
        if self.board[3]==player and self.board[5]==player and self.board[7]==player:
            return True
        #COLUMNS
        for i in range(1,4):
            if self.board[i]==player and self.board[i+3]==player and self.board[i+6]==player: 
                return True
        #ROWS
        for i in range(1,8,3):
            if self.board[i]==player and self.board[i+1]==player and self.board[i+2]==player:
                return True
        return False

    def checkGameState(self):
        self.printBoard()
        
        if self.isWinning(self.player):
            print("Player Wins")
            exit()
        elif self.isWinning(self.computer):
            print("Computer Wins")
            exit()
        elif self.isDraw():
            print("DRAW")
            exit()

    def updatePlayerPosition(self, player, position):
        if self.board[position]==' ':
            self.board[position]=player
            self.checkGameState()
        else:
            print("Can't Insert There")
            self.movePlayer()
        
    def movePlayer(self):
        position=int(input("Enter Position to Enter O\n"))
        self.updatePlayerPosition(self.player, position)

    def moveComputer(self):
        print("Computer's Move")
        bestScore=-float('inf')
        bestMove=0
        for position in self.board.keys():
            if self.board[position]==' ':
                self.board[position]=self.computer
                score=self.minimax(0, -float('inf'), float('inf'), False)
                self.board[position]=' '

                if score>bestScore:
                    bestScore=score
                    bestMove=position
        self.board[bestMove]=self.computer
        self.checkGameState()

    def minimax(self, depth, alpha, beta, isMaximizer):
        if self.isWinning(self.computer):
            return REWARD-depth
        if self.isWinning(self.player):
            return -REWARD+depth
        if self.isDraw():
            return 0
        if isMaximizer:
            bestScore=-float('inf')
            for position in self.board.keys():
                if self.board[position]==' ':
                    self.board[position]=self.computer
                    score=self.minimax(depth+1,alpha, beta, False)
                    self.board[position]=' '

                    if score>bestScore:
                        bestScore=score
                    alpha=max(alpha, score)
                    if alpha>=beta:
                        break
            return bestScore
        else:
            bestScore=float('inf')
            for position in self.board.keys():
                if self.board[position]==' ':
                    self.board[position]=self.player
                    score=self.minimax(depth+1, alpha, beta, True)
                    self.board[position]=' '

                    if score<bestScore:
                        bestScore=score
                    beta=min(beta, score)
                    if alpha>=beta:
                        break
            return bestScore
```
